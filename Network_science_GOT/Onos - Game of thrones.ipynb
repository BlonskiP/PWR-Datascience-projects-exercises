Loading Game of thrones network

```python
import matplotlib.pyplot as plt
import networkx as nx
GAME_OF_THRONES_FILE_PATH_NAME = 'GameOfThrones.txt'
G=nx.read_edgelist(GAME_OF_THRONES_FILE_PATH_NAME,data=True, delimiter='.',create_using=nx.DiGraph)#Docelowo create_using=nx.MultiDiGraph
pos = nx.spring_layout(G, scale=1)
edge_labels = nx.get_edge_attributes(G,'relation')
edge_colors = nx.get_edge_attributes(G,'color')
print(edge_labels)
nx.draw_networkx_edge_labels(G, pos, edge_labels  = edge_labels)
nx.draw_networkx_labels(G,pos, data=None,font_color ='r')
nx.draw_networkx_edges(G,pos,arrows=True, directed=True, connectionstyle='arc3, rad = 0.3')
nx.draw_networkx_nodes(G, pos)
```
