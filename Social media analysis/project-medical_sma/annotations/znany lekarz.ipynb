```python
import pandas as pd
```

```python
df = pd.read_json('znanylekarz.json',lines=True)
df
```

```python
df = df.drop(columns=['_id','id','url','type'])
```

```python
df['label']='Medyczne-diagnoza'
```

```python
def count_words(text):
    split = text.split()
    for word in split:
        word = word.replace(',','').replace('.','')
    split = list(filter(None, split))

    
    return len(split)
df['words']=df['text'].apply( lambda x: count_words(x))
```

```python
df = df[df.words > 25]
df = df[df.words < 75]
```

```python
df.to_csv('znanylekarz_dataset.csv',index=False)
```

```python
import plotly.express as px
gby = df.groupby(by=["label"]).mean()
gby = gby['words'].to_dict()
gby
plot_dict = {'klasa':list(gby.keys()),'ilość słów':list(gby.values())}
fig = px.bar(plot_dict, x='klasa', y='ilość słów',color="ilość słów")
fig.show()
```
