```python
file = "covid19_2020_02_08.json.gz"
```

```python
import pandas as pd
import gzip
import json
```

```python
import json
import gzip
import pandas as pd

with gzip.GzipFile(file, 'r',) as fin:
    data_lan = []
    for line in fin:
        data_lan.append(json.loads(line.decode('utf-8')))

data = pd.DataFrame(data_lan)
```
