```python
import pandas as pd
```

```python
results = pd.read_csv("results.csv")
results
```

```python
results=results.sort_values(by=['fscore macro'],ascending=False)
```

```python
results
```

```python
group = results.drop(["uncover_rate","accuracy"],axis=1).groupby(by=["utility"],as_index=False).mean()
group.plot.bar(x="utility")
```

```python
group = results.drop(["uncover_rate","accuracy"],axis=1).groupby(by=["cls"],as_index=False).mean()
group.plot.bar(x="cls")
```
